```python
# 1 1. Count vowels and consonants in a string using methods.
st="hey toshiba goda"

vowel="aeiou"
vowel_count=0
conso_count=0

for ch in st:
    if ch.isalpha():
        if ch in vowel:
            vowel_count+=1
        else:
            conso_count+=1
print("vowel: ",vowel_count)
print("consonant",conso_count)
```
