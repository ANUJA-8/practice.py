```python
#Understand how to manipulate strings and print them.
def greet(name):
    greeting="Welcome to Python Programming,"
    print(greeting,name)
greet("Rahul!")
```

```python
# Variable Assignment and Types
age=int(25)
height=float(5.9)
favorite_color="blue"
print("Age:", age, "| Type:", type(age))

print("Height:", height, "| Type:", type(height))

print("Favorite Color:", favorite_color, "| Type:", type(favorite_color))
```

```python
# Working with Lists
fruits=["apple", "banana", "cherry", "date", "elderberry"]
print("First fruit:",fruits[0])
print("Last fruit:",fruits[-1])
print("Fruits from index 1 to 2:",fruits[1:3])
```

```python
# Exploring Tuples
person=("Rahul",25,5.9)
print("Age:",person[1])
```

```python
# Understanding Dictionaries
car={"make":"Toyota" , "model":"Camry" , "year":2020 ,"color":"Blue" }
print("Car model:",car["model"])
car["owner"]="Rahul"
print("Updated car dictionary:",car)
```
