```python
#1. The cost of an array is the value of its first element. For example, the cost of [1,2,3] is 1 while the cost of [3,4,1] is 3.
# You need to divide nums into 3 disjoint contiguous subarrays.
# Return the minimum possible sum of the cost of these subarrays.
class Solution:
    def minimumCost(self, nums):
        n = len(nums)
        answer = float('inf')
        
        for i in range(1, n - 1):
            for j in range(i + 1, n):
                cost = nums[0] + nums[i] + nums[j]
                answer = min(answer, cost)
        
        return answer
nums = [10,3,2,1]
solution = Solution()
print(solution.minimumCost(nums))
 
```

```python
def maxConsecBits(arr):
    maxCount, count, prev = 0, 0, -1

    for num in arr:
        
        # If the current number is the same 
        # as the previous number
        if (prev ^ num) == 0:
            count += 1
        else:
            
            # Update max_count and reset count
            maxCount = max(maxCount, count)
            count = 1
        prev = num

    return max(maxCount, count)

if __name__ == "__main__":
    
    arr = [0, 1, 0, 1, 1, 1, 1]
    
    print(maxConsecBits(arr))
```

```python
# Maximum Number of Non-Overlapping Intervals
# Two intervals are considered non-overlapping if one interval ends before or exactly when the other interval begins.
# Input:start = [1, 3, 5, 7] end= [2, 4, 6, 8]
# Intervals:[1,2], [3,4], [5,6], [7,8]
# Output:4

class Intervals:
    def maxinterval(start,end):
        count=0
        for i in range(0,len(start)-1):
            for j in range(0,len(end)-1):
                if end[j]== start[i+1] or start[i+1]== end[j]+1:
                    count+=2
                else:
                    count+=1
            return count
                    
# start=[1, 3, 5, 7]
# end=[2, 4, 6, 8]
start = [3,4,5]
end   = [5,9,8]
Intervals.maxinterval(start,end)
                    
```

```python
# Convert Sorted Array to Binary Search Tree
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
def sortedArrayToBST( nums):
    if not nums:
        return None
    mid = len(nums) // 2        
    root = nums[mid]
    root.left = sortedArrayToBST(nums[:mid])
    root.right = sortedArrayToBST(nums[mid+1:])
        
    return root
nu=[-10,-3,0,5,9]
sortedArrayToBST(nu)
```

```python
#Implement Stack using List
class Stack:
    def __init__(self):
        self.stack=[]
    def push(self,ch):
        self.stack.append(ch)        
    def pop(self):
        if self.stack:
            return self.stack.pop()
        return "stack is empty"
    def peek(self):
        if self.stack:
            return self.stack[-1]
        return "stack is empty"
s = Stack()
s.push(10)
s.push(20)
print(s.pop())   # 20
print(s.peek())
```

```python
# add two numbers given in string dont use inbuilt function 
def addStrings(num1, num2):
        i = len(num1) - 1
        j = len(num2) - 1
        carry = 0
        result = ""

        while i >= 0 or j >= 0 or carry:
            n1 = ord(num1[i]) - ord('0') if i >= 0 else 0
            n2 = ord(num2[j]) - ord('0') if j >= 0 else 0

            total = n1 + n2 + carry
            carry = total // 10

            result = str(total % 10) + result

            i -= 1
            j -= 1

        return result
addStrings("234","23")
addStrings("234","23")
```

```python
# Bse 7
def convertToBase7(num):
        if num == 0:
            return "0"
        
        sign = "-" if num < 0 else ""
        num = abs(num)
        
        result = ""
        
        while num > 0:
            result = str(num % 7) + result
            num //= 7
        
        return sign + result
# convertToBase7(100)
convertToBase7(-10)
```

```python
def reverseVowels(s):
    vowels = set("aeiouAEIOU")
    s = list(s)   # convert string → list (mutable)

    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] not in vowels:
            left += 1
        elif s[right] not in vowels:
            right -= 1
        else:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1

    return "".join(s)
reverseVowels("Arefgw")
    
```

```python
def isSubsequence(s, t):
    i = 0  # pointer for s
    j = 0  # pointer for t

    while i < len(s) and j < len(t):
        if s[i] == t[j]:
            i += 1
        j += 1

    return i == len(s)
# isSubsequence("abp", "ahbgdc")
isSubsequence("abc", "ahbgdc")
```
