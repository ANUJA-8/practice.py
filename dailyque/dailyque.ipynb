```python
#1. The cost of an array is the value of its first element. For example, the cost of [1,2,3] is 1 while the cost of [3,4,1] is 3.
# You need to divide nums into 3 disjoint contiguous subarrays.
# Return the minimum possible sum of the cost of these subarrays.
class Solution:
    def minimumCost(self, nums):
        n = len(nums)
        answer = float('inf')
        
        for i in range(1, n - 1):
            for j in range(i + 1, n):
                cost = nums[0] + nums[i] + nums[j]
                answer = min(answer, cost)
        
        return answer
nums = [10,3,2,1]
solution = Solution()
print(solution.minimumCost(nums))
 
```

```python
#3.Third largest element in an array of distinct elements
class ThirdLarge:
	def third_large(arr):
		arr=list(set(arr))
		arr.sort()
		if len(arr)<3:
			return "no third element exist"
		else:
			return f"Third largest element is {arr[-3]}"

ar=list(map(int,input("array number using space").split()))
ThirdLarge.third_large(ar)
			
```

```python
#4.1 array reverse
class Reverse:
    def rev_arr(arr):
        start=0
        end=len(arr)-1
        while start<end:
            arr[start], arr[end]=arr[end],arr[start]
            start+=1
            end-+1
        return arr
ar=list(map(int,input("Input the array in space").split()))
Reverse.rev_arr(ar)
```

```python
#Reverse an Array in groups of given size
class Reve_group_ar:
	def reve_grp(arr,k):
		result=[]
		for i in range(0,len(arr),k):
			result.extend(arr[i:i+k][::-1])
		return result
#ar=list(map(int,input("Array with space").split()))
ar=[2,4,3,5,6,4,7]
Reve_group_ar.reve_grp(ar,3)
```
