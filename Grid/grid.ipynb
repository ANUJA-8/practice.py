```python
# Count how many 1s exist.
def countOne(grid):
    count=0
    for rows in grid:
        for num in rows:
            if num==1:
                count+=1
    return count
print(countOne([[1,0,0],[0,1,0],[0,0,1]]))
```

```python
# Find Maximum Element of grid
def maxOfGrid(grid):
	max_val=grid[0][0]
    for row in grid:
		for j in row:
			if j>max_val:
				max_val=j
    return max_val
		
def find_max(grid):
    max_val = grid[0][0]
    for row in grid:
        for num in row:
            if num > max_val:
                max_val = num
    return max_val
			
```

```python
# Right Triangle (Increasing Stars)
def right_tri(n):
    for i in range(1, n+1):
        for j in range(i):
            print("*",end=" ")
        print()
right_tri(4)
                
```

```python
# Right Triangle Inverted Right Triangle
def right_top_tri(n):
    for i in range(n,0,-1):
        for j in range(i):
            print("*",end=" ")
        print()
right_top_tri(4)
                
```

```python
# max in grid using column and row
def max_grid(grid):
    m = len(grid)
    print (m)
    n = len(grid[0])
    print (n)
    max_val=grid[0][0]
    for row in range(m):
        for col in range(n):
            if max_val<grid[row][col]:
                max_val=grid[row][col]
    return f"max value is {max_val}"
gr=[
 [1,2,3],
 [4,5,6]
]
max_grid(gr)
```

```python
# convert row col grid in col row grid (Transpos Matrix)
def trans_grid(grid):
    m=len(grid) #counts rows
    n= len(grid[0]) #counts numbers in row
    for row in range(n):
        for col in range(m):
            print(grid[col][row], end="\t")
        print()
trans_grid(gr)
```

```python
#2.sum of column
def column_sum(grid):
    row=len(grid)
    col=len(grid[0])
    sum1=[]
    for j in range(col):
        sum=0
        for i in range(row):
            sum += grid[i][j]
        sum1.append(sum)
    return sum1
gr=[
 [1,2,3],
 [4,5,6]
]
column_sum(gr)


```

```python
#3. count even numbere grid
def eve_count(grid):
    for rows in grid:
        count=0
        for n in rows:
            if n%2==0:
                count+=1
        count+=1
    return count
    # print()
eve_count(gr)
```

```python
#4. Return True if x exists in grid.
def is_element_in_grid(grid,x):
    for rows in grid:
        for n in rows:
            if n==x:
                return True
        return False
gr=[
 [1,2,3],
 [4,5,6]
]
is_element_in_grid(gr,2)
```

```python
#5. Print only border elements.
def boundry_ele(grid):
    r=len(grid)
    c=len(grid[0])
    for i in range(r):
        for j in range(c):
            if i==0 or j==0 or i==r-1 or j==c-1:
                print(grid[i][j])
            else:
                print(" ")   
gr=[
 [1,2,3],
 [4,5,6],
 [6,9,8]
]                                               
boundry_ele(gr)
```
